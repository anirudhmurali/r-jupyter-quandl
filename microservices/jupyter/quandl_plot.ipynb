```r
library(jsonlite)
library(httr)
library(ggplot2)

quandl_data <- '{
    "type": "select",
    "args": {
        "table": "zillow_m1037_zrisfrr",
        "columns": [
            "date",
            "value"
        ]
    }
}';

r <- POST("https://data.chock95.hasura-app.io/v1/query", body = quandl_data)
s <- content(r, "text")
p <- fromJSON(s)
df = data.frame(p$date, p$value)
df
plot(df)
```
